```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
X_org = np.random.normal(0,1,(n,2))
plt.scatter(X_org[:, 0],X_org[:, 1])
plt.show()
```

```python
s=np.array([ [1, 0.8], [0.8, 1] ])
X = np.dot( X_org , s )
print(X.shape)
plt.scatter(X[:, 0],X[:, 1])
plt.show()
```

```python
D, V = np.linalg.eig(s)
print(V)
print(D)
```

```python
newX = X * V[:,1]
```

```python
plt.
plt.hist(X_org, 50)
plt.show()
```

```python
plt.hist(newX, 50)
plt.show()
```
